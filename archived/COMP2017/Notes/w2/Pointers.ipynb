## Pointers

We declare a pointer using the `*` symbol after the type (i.e. `int* ptr`).

The `&` and `*` symbols are useful in relation to pointers:
- `&` gets the address of a variable (i.e. `ptr = &a`).
- `*` dereferences, or gets the value at the address (i.e. `int b = *ptr`).

We can set the dereferenced value to a new value (i.e. `*ptr = 5`). This is going to the address and changing whatever is there to the new value.

Segmentation faults can occur when we try to access memory that we don't have access to.

## Memory

C accesses virtual memory, so we are able to store data contiguously. The physical memory that C internally translates virtual memory into can actually be spread out.

### sizeof

The `sizeof` operator (not a function) returns the number of bytes used to represent the given type or expression.

While it can be used to get the length of an array, it is not recommended. For instance, an array passed into a function decays into a pointer, and will always have `sizeof(void *)`, the size of the pointer rather than the total size of the array.

## Arrays

Arrays are indexed collections of the same type. They are contiguous, meaning each element occurs right after one another in memory.

### Declaration

```c
char alphabet[26];

char vowels[5] = {0}; // initialises all elements to '\0'
```

The array variable `alphabet` is a pointer to the first element in the array.

### Initialisation

```c
for (int i = 0; i < 26; i++) {
    alphabet[i] = 'a' + i; // can perform additions to chars like they're ints
}
```

We can also do `char vowels[] = {'a', 'e', 'i', 'o', 'u'}`, but only when declaring a new array.

**Never** use a variable as the length of an array; the stack is small and your program will crash if you have an array that is too large.

### Strings

In C, strings resemble an array of characters, terminated with a null byte `'\0'`.

When we work with strings, we go through each character until a null byte is read.

Note that there is a difference between `char*` and `char[]` string literals:
- `char*` string literals are allocated in static memory.
- `char ...[]` string literals are allocated in the stack.

## Pointer Arithmetic

We can add and subtract from pointers, which lets us access values that are adjacent to each other in memory.

Pointer arithmetic depends on the size of its data type. So, if we have an `int* arr` with address $\texttt{0x7000}$, `arr + 1` will give $\texttt{0x7004}$ (assuming `sizeof(int)` is 4).

### Pointer Notation

Accessing `arr[1]` is the same as dereferencing `*(arr + 1)`. This also means that `arr + 1` is the same as `&arr[1]`.

---

## Pointer Interpretation

Suppose we have `int** data`. This can be interpreted as:

1. A pointer to a single pointer of an `int` value.
2. An array of addresses that each point to a single `int`.
3. A pointer to an array of `int` values.
4. An array of addresses that point to arrays of `int` values.

## Security

### Uninitialised Values

Uninitialised values can remain uninitialised when we might not expect them to, either by not checking the outcome of an operation, or if the flow control is not correct.

### scanf

If a string longer than the provided width-specifier is read by `scanf`, it will write outside the bounds of the array. This can potentially crash the program, overwrite the values of other variables or introduce a buffer overflow security vulnerability.
