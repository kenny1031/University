## C vs. Java

Concepts learned from programming in Java largely transfer over to C; your functions, loops, variables, arrays, etc. Control flow structures (if, for, while) are also the same.

C lacks garbage collection; all memory allocated by the programmer must be manually freed. There are no classes or objects in C, and references are called "pointers".

Whereas Java is designed around an object-oriented philosophy, C is closer to the underlying machine. It has simple memory model and assumes the programmer knows best.

---

## Compilation

Compilation converts human readable C code to machine level code that can be fed directly to the CPU. 

Compile using GCC (GNU Compiler Collection):
```
gcc hello.c -o hello
```
This will create a binary executable file `./hello`.

### Compiler Flags

Here are some more flags:
|flag              |usage                                     |
|------------------|------------------------------------------|
|`-o <filename>`   |Place compiled binary into file           |
|`-l<library>`     |Add a library (.lib) file                 |
|`-Wall`/`-Wextra` |Enable more warnings                      |
|`-I<dir>`         |Add /dir as a location to search for files|

`-lm` will link math into the binary.

### Compiler Pipeline

When we call `gcc` to compile a file, there are three steps that are done:

1. Preprocessing - text substitution.
2. Compilation/Assembling - convert C code to machine code.
3. Linking - combine multiple files together to one binary.

---

## Preprocessing

### #define

You are able to define constants and macros to be evaluated at compilation using preprocessor directives.
```c
#define PI 3.14
```
Define is a text to text process. Imagine this as replacing `PI` with `3.14` throughout your program.

### #include

The `include` statement allows you to bring in header files which will expose function prototypes your program for you to use.
```c
#include <stdio.h>
```
For local header files, use double quotes around the file.
```c
#include "local.h"
```

## Types

C is statically typed, so all variables must have a type associated with it.

The following types and their respecive sizes are as they occur on 64-bit architectures (which can be obtained by using `sizeof`):
|type                        |bytes                   |
|----------------------------|------------------------|
|`char` and `unsigned char`  |1                       |
|`int` and `unsigned int`    |4                       |
|`short` and `unsigned short`|2                       |
|`long` and `unsigned long`  |4 (windows) or 8 (linux)|
|`double`                    |8                       |
|`float`                     |4                       |
|`void*` (pointers)          |8                       |

Note that how each type is defined differs according to what CPU architecture, platform and compiler that is used. Unless otherwise specified, we cannot assume the size of each data type.

To be safe, always use integer types specified in `<stdint.h>`.

## Less Familiar Types

### signed / unsigned

Most C implementations default types as `signed` values, but a warning that we should not assume this.

Sign usage can be enforced with `signed` and `unsigned` specifiers.

### const

`const` prevents the value being modified, used to prevent arbitrary changes to memory. The value that it protects depends on where it appears:
```c
const char* ptr1; // ptr1 can be changed but not the char it stores
char* const ptr2; // char at ptr2 can be changed but the address cannot
```

### enum

Enums (enumerated types) associate names with integers. For example:
```c
enum day_name {
    Sun, Mon, Tue, Wed, Thu, Fri, Sat, day_undef
};
```
This maps to integers 0 to 7. Much like an `int`, you can do things like `Sun++`.
