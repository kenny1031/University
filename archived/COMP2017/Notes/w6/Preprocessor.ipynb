# Preprocessor

---

## C Preprocessor

You can examine the raw output of the preprocessor by calling it directly:
```
cpp tasks.c
```
Or by instructing the compiler to only perform the preprocessing step (`-E` to dump the output of the preprocessor).
```
gcc -E tasks.c
```

---

## Compiling and Linking

We use the `-c` flag to produce an object file. An object file is written in machine language - assembly in binary format for the target CPU - but is not linked with other parts of your program.
```
gcc -c program.c
```
This will produce `program.o`.

To produce an executable file, we need to link all of our `.c` programs and `.o` files together using `gcc -o` or `ld`.
```
gcc -o program program.o util.c
```
Now we should have the `./program` executable.

---

## Generating Assembly

You can ask the compiler to stop after assembly generation with the `-S` flag (also with the debugging flag `-g` enabled for the assembler).
```
gcc -S -g program.c
```
This will produce `program.s`, the output of the assembler.

You can also extract the assembly from object files with `objdump`.
```
objdump -s program.o
```

## Linkage

### extern

When the `extern` keyword used within a `.c` source file, it notifies the compiler to link to a variable or function of the same name within the global memory space. 
```
extern int global_val; 
```
This will attempt to find the declaration and definition of `global_val` in another linked `.c` file.



### static

Static variables have internal linkage, and are scoped to only the current file. These variables are not exposed to other files.
```
int func() {
    static int static_val = 0;
}
```
Static variables preserve their value even after they go out of scope. They are initialized only once and retain their value between function calls.

## Macros

You can define any replacement string as a preprocessor symbol with `#define`.

Use with caution! Suppose you have declared
```c
#define LINES 10+10
```
And have used
```c
int pagesize = LINES * 5;
```
This will give you a pagesize of 60, **not** 100. Always bracket expressions in defined symbols.

A macro looks like a function with parameters.
```c
#define min(a,b) ((a) < (b) ? (a):(b))
```
Note, this is a ternary operator: `<boolean expression> ? <value if true> : <value if false>`.

## Conditional Inclusion

The preprocessor allows you to select text to be included or not. This can be very useful for debugging.

`#ifdef` and `#ifndef` test if a preprocessor symbol has or has not been defined.

We use `#undef` do undefine and remove an existing macro.

`#if` and `elif` allows for more comprehensive conditional control.

Both `#ifdef` and `#if` can have an `#else`. 

`#endif` goes at the end of every conditional statement.

## Pre-Defined Symbols

### \_\_LINE__

Contains the line number where it is called.

### \_\_FILE__

Contains the name of the file in which it is called.

## Security of Type Limits

### Integer Overflow/Underflow

Understand that all numeric types have a minimum and maximum value that can be represented. If you try to store an integer that exceeds the type limit, this can lead to an integer overflow error.

### Floating Point Inaccuracy

Floats may be inaccurate; `0.1f` sits between two representable floats, and can't be stored with perfect precision. Therefore, you can never test floats for equality.
