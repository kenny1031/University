# Performance

## Amdahl's Law

> "The performance improvement to be gained from some faster mode of execution is limited by the fraction of the time that the faster mode can be used."

The overall speed-up of a program, expressed as $T_0$, the running time before the performance boost, over $T_1$, the running time after the performance boost, is:
$$
\frac{T_0}{T_1} = \frac 1 {(1-p) + \frac p n}
$$
Where $0 \le p \le 1$ is the fraction of the program that can be parallelised, and $n$ is the maximum number of executable threads.

As $n$ tends to infinity, the overall speed-up is bounded to:
$$
\lim_{n\to\infty} \frac 1 {(1-p) + \frac p n} = \frac 1 {1-p}
$$
The closer $p$ is to $1$, the closer the performance boost is to becoming linear.

## Load Balancing

The granularity of parallelism is how frequently threads have to interact with each other.

| Fine-grained parallelism | Coarse-grained parallelism |
|-|-|
| Small amount of computational work between synchronisation stages | Large amount of computational work between synchronisation stages |
| Low computation-to-communication ratio | High computation-to-communication ratio |
| High communication/synchronisation overhead | Low communication/synchronisation overhead |
| Less opportunity for performance improvement | More opportunity for performance improvement |
|  | Harder to load-balance effectively |

The load balancing problem: threads that finish early have to wait for the thread with the largest amount of work to finish. This leads to idle times, which underutilises parallelism.

Dynamic assignment of work, as opposed to static assignment, is less prone to load imbalance.

## Performance Bottlenecks

Some bottlenecks to a program's performance:
- Large segments of sequential parts
- Highly contended locks
- High communication overhead for little computation
- Poor load balancing
- Scalar code (code that processes one data element at a time)

---

## Performance Loss

There are four sources of performance loss: overhead, non-parallelisable computations, idle times, and contention for resources.

There are trade-offs between different sources of performance loss:
1. Communication vs. computation
2. Memory vs. parallelism
3. Overhead vs. parallelism

## Thread Pool

Starting and joining threads can be very expensive and in practice we want to minimise or completely remove recreating threads. A thread pool is a collection of threads that will perform a general computation task.

In a push system, each thread in the thread pool waits until it has been queued work to do by a coordinator. In a pull system, threads self-assign their work from a single incoming work queue.

## Security Overview

### Buffer Overflow

Overflowing a fixed size buffer lets you overwrite addresses in adjacent memory addresses even if they're unrelated to the buffer you are writing into.

To avoid this, do not use unbounded functions like `scanf` or `strcpy` (instead use `fgets` or `strncpy`).

### Stack Smashing

As part of a function call, the return address is pushed onto the stack, which is what the computer uses to determine what code to jump back to after the function finishes.

Arbitrary code execution can be used to replace the return address on the stack with an address to some arbitrary code to execute. Most modern devices have counter measures for this, however.
