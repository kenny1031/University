# Aggregates and Files

---

## Structures

A struct is a collection of existing types, which has constant size in memory and evaluated at compile time. 
```c
struct coordinate {
    int x;
    int y;
};
```
Each field has its own name and can be called by using the `.` operator.
```c
struct coordinate point;
point.x = 0;
point.y = 0;
```
Note that we must include the 'struct' keyword as part of our type name.

A struct cannot recursively refer to itself, but it can store a struct pointer.

---

## Struct Initialisation

Like arrays, structs can be initialised using curly braces which initialise all values.
```c
struct coordinate point = {0};
```
Or, we can initialise each individual field in the order that they are declared in the struct:
```c
struct coordinate point = {1, 2};
```
Or, we can use the property name explicitly.
```c
struct coordinate point = {.x=1, .y=2};
```

---

## Struct Pointers

Trying to pass a struct into a function as an argument will just copy it, so it is useful to store pointers of structs.
```c
struct coordinate* point_ptr = &point;
```
To access attributes, we can avoid having to arduously write `*(point_ptr).x` by using the `->` operator.
```c
point_ptr->x; // dereference the struct point_ptr and get the attribute named x
```

---

## Typedef

We can improve code readability by repurposing a struct to be used as a type:
```c
typedef struct coordinate {
    int x;
    int y;
} Coordinate;
```
Now we do not need to use the 'struct' keyword as part of the type name.
```c
Coordinate point;
```

## Padding

C will pad struct properties with junk bytes if it doesn't align with the processor's memory.
```c
struct c {
    int x;      // 4 bytes
    short s1;   // 2 bytes
    char c1;    // 1 byte
                // 1 byte of PADDING
    float y;    // 4 bytes
};
```
So, assuming padding to multiples of 4 bytes, the size of this struct is 12 bytes.

Alignment depends on CPU architecture, so it is generally hard to tell what the actual size of a given struct is.

## Unions

Unions are similar to a struct, except instead of storing all properties in one block of memory, they are stored exactly in the same location. The size of the union is the size of the largest possible type it can take on.
```c
union number {
    int i;
    float f;
    double d;
};
```
Every attribute within the union shares the same address. Setting the union to one value and attempting to read it as another will give garbage data. It is the programmer's responsibility to keep track of what value the union takes on (i.e. use an enum or a seperate value).

## Bits

In C there are several bitwise operators that allow us to perform logic on bits:
```c
&    // bitwise AND
|    // bitwise inclusive OR
^    // bitwise XOR (exclusive OR)
~    // bitwise NOT 
```
There are also shift operators that move every bit to the left or right:
```c
<<    // left shift
>>    // right shift
```

---

## Bit Fields

For some specialised applications, you could need data fields that are smaller than a byte, or are packed into several bytes.

In a bitfield, you can specify a size, in bits, for each element of a structure.
```c
struct IOdev {
    unsigned R_W: 1;    // size is placed after the field name, with a colon between
    unsigned Dirn: 8;
    unsigned mode: 3;   // this variable occupies only 3 bits
};
```
Bitfields are good for low level programming (for "unpacking" data structures), but may not be portable due to memory alignment differences on other architectures.

Without using the C bitfield, it is still possible to unpack bitfields by using bitwise shift and logical operations.

## Files

The `<stdio.h>` library contains many standard I/O functions and definitions for using files:
- To open a file, we use the `fopen` function.
- To read from a file, use `fscanf`/`fgets` (generally `fgets` is preferred, especially with line buffered data).
- Write text to the file using `fprintf`.
- For binary data, use `fread` and `fwrite`.
- Finally, don't forget to close the file using `fclose`.

Special files `stdin`, `stdout` and `stderr` are opened already when your program begins.

`feof` is important for testing if the end of file has been reached while reading input files.

`fflush` force writes all data for output streams, and discards unprocessed buffered data for input streams.

## Header Files

While C source code uses the .c extension, we import .h files in our `#include` statement. Including these header files tells our program that these function definitions exist, and the linker provides the code that implements them.

We write a header guard using `#ifdef` and `#ifndef` to prevent a file from including multiple copies of the same header.
```c
#ifndef <FILENAME>_H
#define <FILENAME>_H

// header file goes here...

#endif
```
